```python
matriz3d = np.random.randint(low=1, high=10, size = (2,20,20))
matriz3d
```

```python
matriz_1 = matriz3d.flatten()
n = int(matriz_1.__len__()/2)
matrizX = matriz_1[0:n]
matrizY = matriz_1[n:]
matrizY
```

```python
sns.scatterplot(x=matrizX, y = matrizY)
plt.xticks(np.linspace(start = 0, stop =10, num = 7))
plt.yticks(np.linspace(start = 0, stop =10, num = 5))
plt.show()
```

```python
sns.relplot(x=matrizX, y = matrizY, kind ="line")
plt.xticks(np.linspace(start = 0, stop =10, num = 7))
plt.yticks(np.linspace(start = 0, stop =10, num = 5))
plt.show()
```
